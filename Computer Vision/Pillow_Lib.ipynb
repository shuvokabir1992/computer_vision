```python
from PIL import Image
from PIL import ImageOps
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = Image.open("Ador Bhuna.JPG")
plt.imshow(image)
image.format
image.size
image.mode
```

```python
image_gray = ImageOps.grayscale(image)
plt.imshow(image_gray)
image_gray.mode
image_gray.save("Adu.png")
```

```python
image_gray.quantize(2)
plt.imshow(image_gray)
image_gray.save("Adu.png")
image_gray.save("AduQ.png")
```

```python
red, green, blue = image.split()
plt.imshow(red)
red.save("AdorR.jpg")
```

```python
red, green, blue = image.split()
plt.imshow(green)
green.save("AdorG.jpg")
```

```python
import cv2
imagecv = cv2.imread("Ador Bhuna.jpg")
type(imagecv)
imagecv.shape
plt.imshow(imagecv)
```

```python
new_image = cv2.cvtColor(imagecv, cv2.COLOR_BGR2RGB)
plt.imshow(new_image)
```

```python
image_gray = cv2.cvtColor(imagecv, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray)
cv2.imwrite("Ador_Grey.jpg", image_gray)
```

```python
im_grey = cv2.imread('Ador_Grey.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(im_grey)
```
